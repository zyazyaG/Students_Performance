# Predicting Students' Performance: Data Cleaning

## Business Problem

Academic achievement is crucial in educational environment, as it is important indicator of an educational institution’s performance. Early prediction of students’ performance can help educational institutions to navigate students to the right direction, provide the needed actions at the right moment, and to plan the appropriate training to improve students’ success rate. 

### Data Cleaning

In this notebook I will **explore** what **datasets** I have to work with and what are the **features**. In addition, I will **clean the datasets**, specifically, I will **select and transform features** needed for project purposes, **remove duplicates, manage missing values and merge datasets.**

### Goal of the Data Cleaning

The main goal of the project is to be able to **predict the outcome of the students in the course**, in order to help those who are at risk of failing. I will **analyse the datasets** provided and **prepare the features for EDA Analysis**. At the end of the notebook I will **save new dataframe in data folder**.


## Data 

The datasets for this project were obtained from **Open University Learning Analytics**. It contains data about **courses, students' demographics,** and their **interactions with Virtual Learning Environment** (VLE) for seven selected courses (called modules). The dataset consists of tables connected through **unique identifiers**. All tables are stored in the csv format. The tables available are following:
-	cources.csv
-	assessments.csv
-	vle.csv
-	studentsInfo.csv
-	studentsRegistration.csv
-	studentAssessment.csv
-	studentVle.csv


**Link to Datasets:** https://analyse.kmi.open.ac.uk/open_dataset#description

## Join All Tables

In order to proceed with modeling, I will need to merge all necessary features into one dataset. I will join each table based on relevant keys.

### Join Student Assessments and Assessments Tables

Now that the necessary assessments are chosen for modeling, I will start joining tables. First, I will join stud_assessment to assessment_df

```python
assessment_df.id_assessment.duplicated().sum()
```

```python
stud_assessment.id_assessment.duplicated().sum()
```

```python
df_1 = pd.merge(assessment_df, stud_assessment, how="inner", on=["id_assessment"])
df_1.info()
```

```python
df_1.head(20)
```

```python
group_1 = df_1.groupby(['code_module', 'code_presentation'])
group_1.first()
```

### Join Student Registration Table to DF

```python
df_2 = pd.merge(df_1, stud_reg, how="inner", on=["code_module", "code_presentation", "id_student"])
df_2.info()
```

### Join Student Vle Table to DF

```python
stud_vle.date.value_counts()
```

```python
stud_vle.head(20)
```

#### Select students' vle clicks before 135th day

```python
stud_vle_df = stud_vle[(stud_vle['date'] <= 135)]
stud_vle_df.info()
```

```python
stud_vle_df.head(20)
```

### Sum Up number of click per student's course and presentation

Since students vle table contains number of clicks per student's online application visit, I will sum up those numbers per student's code module and code presentation.

```python
id_stud = stud_vle_df.groupby(['id_student', 'code_module','code_presentation'])
id_stud['sum_click'].sum()
```

```python
id_stud.head()
```

```python
sum_of_click = id_stud['sum_click'].sum()
sum_of_click
```

```python
stud_vle_df = pd.DataFrame(sum_of_click)
stud_vle_df.head()
```

```python
stud_vle_df.reset_index(inplace = True)
```

```python
stud_vle_df.head()
```

```python
stud_vle_df.info()
```

### Join tables

```python
df_3 = pd.merge(df_2, stud_vle_df, how="inner", on=["code_module", "code_presentation", "id_student"])
df_3.info()
```

```python
df_3.head(20)
```

### Join Student Info Table and DF

```python
stud_info.info()
```

```python
df_4 = pd.merge(df_3, stud_info, how="inner", on=["code_module", "code_presentation", "id_student"])
df_4.info()
```

```python
df_4.head()
```

### Drop Observation with Unregistered Students

Since the outcome of students who unregistered from the class is already determined, I will drop those observations from dataframe

```python
df_4 = df_4[df_4.date_unregistration.isna()]
df_4.info()
```

```python
df_4.drop(columns = "date_unregistration", inplace = True)
df_4.info()
```

```python
df_4.head(20)
```

### Calculate mean score for the assessments in each course per student

```python
module_score = df_4.groupby(['id_student', 'code_module', 'code_presentation'])["score"].mean()
```

```python
module_score = pd.DataFrame(module_score)
module_score.head()
```

```python
module_score.reset_index(inplace = True)
module_score.head()
```

```python
module_score.rename(columns = {"score": "mean_score"}, inplace = True)
```

### Join Scores to DF

```python
df_5 = pd.merge(df_4, module_score, how="inner", on=["code_module", "code_presentation", "id_student"])
df_5.info()
```

```python
df_5.head()
```

```python
df_5.groupby(['id_student', 'code_module', 'code_presentation']).mean()[['mean_score']]
```

```python
df_5.sort_values(by = 'num_of_prev_attempts', inplace = True)
```

### Create Final DF and Drop Duplicated Observations 

```python
final_df = df_5.drop_duplicates(['id_student', 'code_module', 'code_presentation'], keep = 'last')
final_df.head()
```

### Join Cources Dataframe

```python
final_df.info()
```

```python
courses.head()
```

```python
final_df = pd.merge(final_df, courses, how="inner", on=["code_module", "code_presentation"])
final_df.info()
```

## Clean Final Dataframe

* Drop columns that ar relevant to assessments - date, weight an date of submision.

```python
final_df.drop(columns = ['date', 'weight', 'date_submitted', 'score'], inplace = True)
final_df.head()
```

```python
final_df.isna().sum()
```

```python
final_df.imd_band.value_counts(normalize = True)
```

```python
final_df.imd_band.value_counts()
```

* Drop observations with missing imd_band feature values

```python
final_df.dropna(subset = ['imd_band'], inplace = True)
```

```python
final_df.info()
```

### Remove Features

I will remove features that are not relevant to the project and modeling purpose 

```python
final_df.drop(columns = ['id_assessment', 'assessment_type', 
                         'is_banked'], inplace = True)
final_df.info()
```

```python
final_df.head(20)
```

### Create Row Index

In order to make each observation uniquely identified, I will create a column that will be combination of code module, code presentation and student id.

```python
final_df['index_row'] = final_df.code_module + final_df.code_presentation + final_df.id_student
final_df.head()
```

```python
cols = list(final_df.columns)
cols = [cols[15]] + cols[0:15]
final_df = final_df[cols]
```

```python
final_df['date_registration'] = final_df['date_registration'].astype(int).abs()
final_df.head()
```

```python
final_df.reset_index(inplace = True, drop = True)
final_df.head(10)
```
